```python
from matplotlib import pyplot as plt
import numpy as np
from utils.hypothesis import *
from utils.data import *
```

```python
def calculs_with_saf_per_year(annee, #int : année à calculer
                     incorpo,         # float : taux d'incorporation (hypothèse)
                     allowance_free_2023, # float : allowance d'émissions gratuites en 2023 (hypothèse)
                     price_saf,           # float : prix du SAF au litre (hypothèse)
                     price_kero,          # float : prix du kérosène au litre (hypothèse)
                     allowance_SAF        # float : allowance gratuite de réduction du surcoût lié au SAF (hypothèse)
                     ):
    # Recupère les données
    y = annee-2023
    carbonprice_one_year = carbonprice[y]
    quota_eu_one_year =  quota_eu[y]
    volume_carbu_an_one_year = volume_carbu_an[y]
    
    # Cas sans SAF : 
    prix_kero_an = volume_carbu_an_one_year * price_kero
    prix_carbone_kero_an = (CO2_em - allowance_free_2023*quota_eu_one_year) * carbonprice_one_year

    # Cas avec SAF :
    y = annee - 2023
    volume_saf_an = volume_carbu_an_one_year*incorpo
    volume_kero_saf_an = volume_carbu_an_one_year*(1-incorpo)

    # Calcul cout carbone kérosène
    CO2_em_kero = volume_kero_saf_an * nrj_volum_kero * core_lca_kero /1000000
    prix_carbone_kero_saf = ( CO2_em_kero - allowance_free_2023*quota_eu_one_year) * carbonprice_one_year

    # Calcul surcoût SAF
    prix_saf_an = volume_saf_an * price_saf
    prix_kero_saf_an = volume_kero_saf_an * price_kero

    prix_total_an = prix_saf_an + prix_kero_saf_an

    prix_extra_an =  prix_total_an - prix_kero_an

    allowances_extra_an = -prix_extra_an * allowance_SAF

    prix_ap_allowances_an = prix_total_an + allowances_extra_an
    
    tot = prix_carbone_kero_saf + prix_ap_allowances_an

    return tot
```

```python
def calculs_with_saf_per_year_heat(annee, #int : année à calculer
                     incorpo,         # float : taux d'incorporation (hypothèse)
                     allowance_free_2023, # float : allowance d'émissions gratuites en 2023 (hypothèse)
                     price_saf,           # float : prix du SAF au litre (hypothèse)
                     price_kero,          # float : prix du kérosène au litre (hypothèse)
                     allowance_SAF,       # float : allowance gratuite de réduction du surcoût lié au SAF (hypothèse)
                     carbonpricegiven,   #float
                     quotagiven
                     ):
    # Recupère les données
    y = annee-2023
    carbonprice_one_year = carbonpricegiven
    quota_eu_one_year =  quotagiven
    volume_carbu_an_one_year = volume_carbu_an[y]
    
    # Cas sans SAF : 
    prix_kero_an = volume_carbu_an_one_year * price_kero
    prix_carbone_kero_an = (CO2_em - allowance_free_2023*quota_eu_one_year) * carbonprice_one_year

    # Cas avec SAF :
    y = annee - 2023
    volume_saf_an = volume_carbu_an_one_year*incorpo
    volume_kero_saf_an = volume_carbu_an_one_year*(1-incorpo)

    # Calcul cout carbone kérosène
    CO2_em_kero = volume_kero_saf_an * nrj_volum_kero * core_lca_kero /1000000
    prix_carbone_kero_saf = ( CO2_em_kero - allowance_free_2023*quota_eu_one_year) * carbonprice_one_year

    # Calcul surcoût SAF
    prix_saf_an = volume_saf_an * price_saf
    prix_kero_saf_an = volume_kero_saf_an * price_kero

    prix_total_an = prix_saf_an + prix_kero_saf_an

    prix_extra_an =  prix_total_an - prix_kero_an

    allowances_extra_an = -prix_extra_an * allowance_SAF

    prix_ap_allowances_an = prix_total_an + allowances_extra_an
    
    tot = prix_carbone_kero_saf + prix_ap_allowances_an

    return tot
```

```python
def heatmap_calculs_with_saf(annee,                # int : année à calculer
                              incorpo,              # float : taux d'incorporation (hypothèse)
                              allowance_free_2023,  # float : allowance d'émissions gratuites en 2023 (hypothèse)
                              price_saf_list,       # list : liste des prix du SAF au litre
                              price_kero,           # float : prix du kérosène au litre (hypothèse)
                              allowance_SAF,        # float : allowance gratuite de réduction du surcoût lié au SAF (hypothèse)
                              carbonprice_list,     # list : liste des prix du carbone
                              quota_eu_list         # list : liste des quotas européens
                             ):

    heatmap_data = np.zeros((len(quota_eu_list), len(price_saf_list)))

    for i, carbonprice_one_year in enumerate(carbonprice_list):
        for j, price_saf_one_year in enumerate(price_saf_list):
            tot = calculs_with_saf_per_year_heat(annee, incorpo, allowance_free_2023, price_saf_one_year,
                                                  price_kero, allowance_SAF, carbonprice_one_year, quota_eu_list[0])
            heatmap_data[j, i] = tot

    # Plot heatmap
    plt.imshow(heatmap_data, cmap='viridis', extent=[min(carbonprice_list), max(carbonprice_list),
                                                     min(price_saf_list), max(price_saf_list)],
               origin='lower', aspect='auto')
    
    plt.colorbar(label='total costs')
    plt.xlabel('Carbon Price')
    plt.ylabel('Price SAF')
    plt.suptitle(f'Total costs in function of SAF and Carbon Price')
    plt.title(f'for an incorporation rate of {incorpo} in {annee}')
    plt.show()
```

```python
carbonprice_list = np.linspace(80, 150, 10)
price_saf_list = np.linspace(1.0, 4.0, 10)  # Remplacez par la plage de valeurs réelle

heatmap_calculs_with_saf(2024, 0.2, allowance_free_2023, price_saf_list, price_kero,
                         allowance_SAF, carbonprice_list, quota_eu_list)
```
